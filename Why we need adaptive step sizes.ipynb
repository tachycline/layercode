```python
import numpy as np
from scipy.integrate import odeint
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

def fp(f, t):
    return 9.8-.004*f**2 + t**2

def ff(f0, t0, h):
    k1 = fp(f0, t0)
    f1 = f0 + k1 * h / 2
    k2 = fp(f1, t0+h/2)
    f2 = f0 + k2 * h / 2
    k3 = fp(f2, t0+h/2)
    f3 = f0 + k3 * h
    k4 = fp(f3, t0+h)
    m = k4/6 + k3/3 + k2/3 + k1/6
    return f0 + m*h
```

```python


step = 11.0
tmax = 20.0
tt = np.linspace(0.0, tmax, 51)

tta = np.linspace(0.0, tmax, step)
fa = np.zeros(tta.size)
f0=0
fa[0]=f0
h = tmax / (step - 1)
for idx, tval in enumerate(tta):
    if idx < len(tta)-1:
        fa[idx+1] = ff(fa[idx], tval, h)
```

```python
plt.plot(tta, fa)
plt.plot(tt, odeint(fp, f0, tt))
plt.ylabel('f(t)')
plt.xlabel('t')
```
